### **Naive Bayes Classifier**



**What is Naive Bayes algorithm?**

Naive Bayes is a classification technique that is based on Bayes' Theorem (*Probability theory*) and assumes that all of the features that predict the target value are independent of one another (Yang, 2018). In determining the target value, a Naive Bayes classifier assumes that the presence of a particular feature in a class is unrelated to the presence of any other feature.

Nave Bayes algorithm is a supervised learning algorithm, which is based on Bayes theorem and used for solving classification problems.

-> **Nave:** It is called Nave because it assumes that the occurrence of a certain feature is independent of the occurrence of other features.

-> **Bayes:** It is called Bayes because it depends on the principle of Bayes' Theorem.

Naive Bayes is a popular algorithm for natural language processing (NLP) tasks such as spam filtering, sentiment analysis, classification, and recommendation (Pawel, 2015). It is founded on *Bayes' Theorem*, as illustrated below.

>
> $$\mathbf{P} \left({x \mid c} \right) = \frac{\mathbf{P} \left ({c \mid x} \right) \mathbf{P} \left({c} \right)}{\mathbf{P} \left( {x} \right)}$$

In the *above* equation,

* P(c|x) is the posterior probability of class (c, target) given predictor (x, attributes).
* P(c) is the prior probability of class **c**.
* P(x|c) is the likelihood which is the probability of predictor(the query  **x**) given class.
* P(x) is the prior probability of predictor **x**

The Bayes' theorem is used by the Nave Bayes Classifier to predict membership probabilities for each class, such as the likelihood that a given record or data point belongs to a specific class. The most likely class is the one with the greatest probability. This is refered to as **Maximum A Posteriori (MAP)**.

The **MAP for a hypothesis with 2 events A and B is**

**MAP (A)**

= max (P (A | B))

= max (P (B | A) * P (A))/P (B)

= max (P (B | A) * P (A))

Here, P (B) is evidence probability. It is used to normalize the result. It remains the same, So, removing it would not affect the result.

The Nave Bayes Classifier presumes that all features are unrelated to one another. The presence or absence of one feature has no effect on the presence or absence of another.

We test a hypothesis given multiple evidence on features in real-world datasets. As a result, the calculations become quite complex. To make things easier, the feature independence approach is used to decouple multiple pieces of evidence and treat them as separate entities.

As a result, the Naive Bayes classifier is a generative classifier, as it computes predictions by modeling each class rather than discriminating between predictions using a decision boundary, as other machine algorithms do.

### **Categorical Naive Bayes**

The Naive Bayes classifier is used to classify categorical features. The categorical Naive Bayes classifier is appropriate for categorical distribution of discrete features. Each feature's categories are drawn from a categorical distribution.

CategoricalNB is a categorical naive Bayes algorithm implementation for categorically distributed data. It is assumed that each feature denoted by the index *`i`* has its own categorical distribution.

For each feature *`i`* in the training set *`X`* , CategoricalNB estimates a categorical distribution for each feature *`i`* of *`x`* conditioned on the class *`y`*. The index set of the samples is defined as $$J = \{ 1, \dots, m \}$$, with *`m`* as the number of samples.

The probability of category *`t`* in feature *`i`* given class *`t`* is estimated as:

>
>$$P(x_i = t \mid y = c \;\ \alpha) = \frac{ N_{tic} + \alpha}{N_{c} +
                                       \alpha n_i}$$

where $$N_{tic} = |\{j \in J \mid x_{ij} = t, y_j = c\}|$$ is the number of times category *`t`* appears in the samples  *`xi`* which belong to class *`c`*, $$N_{c} = |\{ j \in J\mid y_j = c\}|$$ is the number of samples with class c,  is a smoothing parameter and *`ni`* is the number of available categories of feature *`i`*.

**NB:** CategoricalNB assumes that the sample matrix  is encoded (for instance with the help of OrdinalEncoder).

### **Random Forest Classifier**


`Random Forest` is an `ensemble learning-based` supervised machine learning algorithm. It comes in two variations: one for `classification problems` and one for `regression problems`. It is one of the most adaptable and user-friendly algorithms.
It constructs multiple decision trees from the given data samples, obtains predictions from each tree, and votes to determine the best solution. It's also a good predictor of feature importance.

`Decision Trees work` by splitting using the variable that produces the purest child node. The varibale selected is based on `Information Gain`.
Information gain is a function that uses either entropy or gini to calculate of the parent and the resulting child nodes.

Information_Gain.PNG


`Entropy` is a measure of disorder. For our model we used entropy but the default for the function is gini.

Decision Tress the variables are  split into two child nodes until all the resulting pure child nodes are obtained.

`Variables are ranked` in-orders of how good they are at predicting the target label with the varibale providing the highest information gain placed at the top.

`Why  Random Forest and not Decision Trees'
1. A decision Tree has high variance. This means thats small changes in the training data will affect the results
2. One major disadvantage of decsion Trees is that it is a `gredy algorithm`. It only think about the variable to split and never about the final result. Due to the greediness, it is difficult to find an optimal decion Trees as we cannot trusr the result of any tree thus, we build multiple trees and combine the results to make a prediction. The ensemble method we choose that can constrsut multiple trees is called is Random Forest.



The `Random Forest approach` integrates multiple decision trees, creating a forest of trees, hence the name. The greater the number of trees in the forest, the higher the accuracy of the random forest classifier.

There are two stages of random forest algorithm intuition.
The random forest is built in the first stage by randomly selecting `k` features from a total of `m` features.
To make sure the trees are different from each other only a small random potion is considered. Otherwise training the same number of trees multiple times will not make a difference
 In the first stage, we will do the following: -

1.   Choose k features at random from a total of m features, where `k<m`.
2.   Calculate the node d using the best split point among the `k` features.
3.   Using the best split, divide the `node` into `daughter nodes`.
4.   Repeat `steps 13` until the desired number of nodes is reached.
5.   Create a forest by repeating `steps 1` through `4` `n times` to create `n   trees`.

In the second stage, we use the trained random forest algorithm to make predictions.

1.   We take the test features and predict the outcome using the rules of each randomly generated decision tree, and we save the predicted outcome.
2.   The votes for each predicted target are then calculated.
3.   Finally, we use the highest voted predicted target as the random forest algorithm's final prediction.

### Categorical columns casting via `asType` into `pd.Categorical` type

This section contains three functions for detecting categorical features, gathering information about them, and finally converting their type to categorical.

```python
#Functions to check for categorical features and then convert the features

def find_cat_columns(dataset_df, cut_off=10):
    """
        This Function is used to search for categorical features in the dataset
    """
    categorical_columns = []
    for c in dataset_df.columns:
        if len(dataset_df[c].unique()) <= cut_off:
            categorical_columns.append(c)

    return categorical_columns

def categorical_converter(cat_columns, dataset_df):
    """
        This Function is used to convert categorical features type into pd.categorical
    """
    for c in cat_columns:
        dataset_df[c] = dataset_df[c].astype('category')

    return dataset_df

def cat_info(dataset_df, unique_types=False):
    """
       This function Gather info about each column feature
    """
    res = [[dataset_df[col].unique(), len(dataset_df[col].unique()),
            dataset_df[col].isnull().sum()] for col in dataset_df.columns]
    dataset_df_temp = pd.DataFrame(res, index=dataset_df.columns,
                           columns=['Uniques', 'No. of uniques', 'No. of NaN'])

    return dataset_df_temp.iloc[:, 0 if unique_types else 1:]
```

##  Calculation of Qualitative features correlation
To calculate `correlation` between `categorical variable`(`Nominal features`) it's better to use `Cramer's V` which is better in terms of performance compared to  `Pearson correlation` approach.

`Cramer's V` is a tool for determining the strength of a correlation between two variables. The interested variables ought to be `categorical`, with two or more unique values per category, in order to use it. More information is provided below.

Hence the `Cramer's V` approach is used under the following:

```python
#The cat_corr and cramers_func are used to used to calculate the correlations between the categorical features

def cat_corr(dataset_df):
    """
        This functions transforms the categorical features into a matrix which the cramers_func function can use and calculate the correlationss
    """
    #Selecting categorical features
    dataset_df = dataset_df.select_dtypes(include='category')
    cols = dataset_df.columns
    n = len(cols)
    corr_matrix = pd.DataFrame(np.zeros(shape=(n, n)), index=cols, columns=cols)

    for col1 in cols:
        for col2 in cols:
            if col1 == col2:
                corr_matrix.loc[col1, col2] = 1
                break
            df_crosstab = pd.crosstab(dataset_df[col1], dataset_df[col2], dropna=False)
            corr_matrix.loc[col1, col2] = cramers_func(df_crosstab)

    # Flip and add to get full correlation matrix
    corr_matrix += np.tril(corr_matrix, k=-1).T
    return corr_matrix

def cramers_func(matrix_table):
    """
        The function Calculates correlation values between categorical features based on Cramer's approach
    """
    chi2 = chi2_contingency(matrix_table)[0]
    n = matrix_table.sum().sum()
    phi2 = chi2/n

    r, k = matrix_table.shape
    r_corrected = r - (((r-1)**2)/(n-1))
    k_corrected = k - (((k-1)**2)/(n-1))
    phi2_corrected = max(0, phi2 - ((k-1)*(r-1))/(n-1))

    return (phi2_corrected / min( (k_corrected-1), (r_corrected-1)))**0.5
```

###  Implementation of One-hot Encoding and Standardization of the Whole dataset

For most machine learning models to perform well it's necessary for the numerical features to be scaled. This keeps the model  from learning to well from features with lower variance. Categorical variables do not need to be standardised.

In addation, it's better to standardize the X_split only because the _test data should be new to the models. This is done to avoid data leak to the testing test.

```python
def data_split(x,y, testSize=.25, rand=0, shuf=True, strat=y):
  data_splits = train_test_split(x, y, test_size=0.25, random_state=0,
                                shuffle=True, stratify=strat)
  x_train, x_test, y_train, y_test = data_splits


  # Naive Bayes
  data_splits2 = train_test_split(x, y, test_size=0.25, random_state=0,
                                shuffle=True, stratify=y)
  x_train_cat, x_test_cat, y_train_cat, y_test_cat = data_splits


  # Save the non-scaled version of monthlycharges and totalcharges to compare classifiers
  x_test_charges = np.array(x_test[['monthlycharges', 'totalcharges']], copy=True)

  list(map(lambda x: x.shape, [x, y, x_train, x_test, y_train, y_test]))

  #########################################################
  categorical_columns = list(x.select_dtypes(include='category').columns)


  ## Column Transformer
  transformers = [('one_hot_encoder',
                    OneHotEncoder(drop='first',dtype='int'),
                    categorical_columns),
                  ('MinMax_Scaler', MinMaxScaler(), numeric_columns)]
  x_trans = ColumnTransformer(transformers, remainder='passthrough')

  # Applying Column Transformer
  x_train = x_trans.fit_transform(x_train)
  x_test = x_trans.transform(x_test)
  new_x = x_trans.fit_transform(x)

  ## Label encoding
  y_trans = LabelEncoder()
  y_train = y_trans.fit_transform(y_train)
  y_test = y_trans.transform(y_test)


  ## Save feature names after one-hot encoding for feature importances plots
  feature_names = list(x_trans.named_transformers_['one_hot_encoder'] \
                              .get_feature_names(input_features=categorical_columns))
  feature_names = feature_names + numeric_columns

  return x_train, x_test, y_train, y_test, x_test_charges, feature_names
```

```python
def confusion_plot(matrix, labels=None):

    labels = labels if labels else ['Negative (0)', 'Positive (1)']

    fig, ax = plt.subplots(nrows=1, ncols=1)
    sns.heatmap(data=matrix, cmap='Blues', annot=True, fmt='d',
                xticklabels=labels, yticklabels=labels, ax=ax)
    ax.set_xlabel('PREDICTED')
    ax.set_ylabel('ACTUAL')
    ax.set_title('Confusion Matrix')
    plt.close()

    return fig
```

```python
def roc_plot(y_true, y_probs, label, compare=False, ax=None):

    fpr, tpr, thresh = roc_curve(y_true, y_probs, drop_intermediate=False)
    auc = round(roc_auc_score(y_true, y_probs), 2)

    fig, axis = (None, ax) if ax else plt.subplots(nrows=1, ncols=1)
    label = ' '.join([label, f'({auc})']) if compare else None
    sns.lineplot(x=fpr, y=tpr, ax=axis, label=label)

    if compare:
        axis.legend(title='Classifier (AUC)', loc='lower right')
    else:
        axis.text(0.72, 0.05, f'AUC = { auc }', fontsize=12,
                  bbox=dict(facecolor='green', alpha=0.4, pad=5))

        # Plot No-Info classifier
        axis.fill_between(fpr, fpr, tpr, alpha=0.3, edgecolor='g',
                          linestyle='--', linewidth=2)

    axis.set_xlim(0, 1)
    axis.set_ylim(0, 1)
    axis.set_title('ROC Curve')
    axis.set_xlabel('False Positive Rate [FPR]\n(1 - Specificity)')
    axis.set_ylabel('True Positive Rate [TPR]\n(Sensitivity or Recall)')

    plt.close()

    return axis if ax else fig
```

```python
def precision_recall_plot(y_true, y_probs, label, compare=False, ax=None):

    p, r, thresh = precision_recall_curve(y_true, y_probs)
    p, r, thresh = list(p), list(r), list(thresh)
    p.pop()
    r.pop()

    fig, axis = (None, ax) if ax else plt.subplots(nrows=1, ncols=1)

    if compare:
        sns.lineplot(r, p, ax=axis, label=label)
        axis.set_xlabel('Recall')
        axis.set_ylabel('Precision')
        axis.legend(loc='lower left')
    else:
        sns.lineplot(thresh, p, label='Precision', ax=axis)
        axis.set_xlabel('Threshold')
        axis.set_ylabel('Precision')
        axis.legend(loc='lower left')

        axis_twin = axis.twinx()
        sns.lineplot(thresh, r, color='limegreen', label='Recall', ax=axis_twin)
        axis_twin.set_ylabel('Recall')
        axis_twin.set_ylim(0, 1)
        axis_twin.legend(bbox_to_anchor=(0.24, 0.18))

    axis.set_xlim(0, 1)
    axis.set_ylim(0, 1)
    axis.set_title('Precision Vs Recall')

    plt.close()

    return axis if ax else fig
```

```python
def feature_importance_plot(importances, feature_labels, ax=None):
    fig, axis = (None, ax) if ax else plt.subplots(nrows=1, ncols=1, figsize=(5, 10))
    sns.barplot(x=importances, y=feature_labels, ax=axis)
    axis.set_title('Feature Importance Measures')

    plt.close()

    return axis if ax else fig
```

```python
def train_clf(clf, x_train, y_train, sample_weight=None, refit=False):
    train_time = 0

    try:
        if refit:
            raise NotFittedError
        y_pred_train = clf.predict(x_train)
    except NotFittedError:
        start = timeit.default_timer()

        if sample_weight is not None:
            clf.fit(x_train, y_train, sample_weight=sample_weight)
        else:
            clf.fit(x_train, y_train)

        end = timeit.default_timer()
        train_time = end - start

        y_pred_train = clf.predict(x_train)

    train_acc = accuracy_score(y_train, y_pred_train)
    return clf, y_pred_train, train_acc, train_time
```

```python
def model_memory_size(clf):
    return sys.getsizeof(pickle.dumps(clf))
#in-built fnction in pickle to convert a python object hieracy in a byte stream
```

```python
def report(clf, x_train, y_train, x_test, y_test, sample_weight=None,
           refit=False, importance_plot=False, confusion_labels=None,
           feature_labels=None, verbose=True):

    dump = dict()

    ## Train if not already trained
    clf, train_predictions, \
    train_acc, train_time = train_clf(clf, x_train, y_train,
                                                     sample_weight=sample_weight,
                                                     refit=refit)
    ## Model execution via cross-validation
    cv_score =  np.mean(cross_val_score(clf, x_train, y_train, scoring='recall',cv=5))

    ## Testing
    start = timeit.default_timer()
    test_predictions = clf.predict(x_test)
    end = timeit.default_timer()
    test_time = end - start

    test_acc = accuracy_score(y_test, test_predictions)
    test_re = recall_score(y_test, test_predictions)
    test_pre = precision_score(y_test, test_predictions)
    test_f1 = f1_score(y_test, test_predictions)

    y_probs = clf.predict_proba(x_test)[:, 1]

    roc_auc = roc_auc_score(y_test, y_probs)


    ## Model Memory
    model_mem = round(model_memory_size(clf) / 1024, 2)

    print(clf)
    print("\n=============================> TRAIN-TEST RESULTS DETAILS <======================================")

    ## Metrics
    print(f"Train Size: {x_train.shape[0]} samples")
    print(f" Test Size: {x_test.shape[0]} samples")
    print("------------------------------------------")
    print(f"Training Time: {round(train_time, 3)} seconds")
    print(f" Testing Time: {round(test_time, 3)} seconds")
    print("------------------------------------------")
    print("Train Accuracy: ", train_acc)
    print(" Test Accuracy: ", test_acc)
    print("------------------------------------------")
    print("------------------------------------------")
    print(" Test Recall: ", test_re)
    print("------------------------------------------")
    print(" Test Precision: ", test_pre)
    print("------------------------------------------")
    print(" Test F1-Score: ", test_f1)
    print("------------------------------------------")
    print(" Area Under ROC: ", roc_auc)
    print("------------------------------------------")
    print(f"Model Memory Size: {model_mem} kB")

    print("\n=============================> CROSS-VALIDATION REPORT <===================================")
    print("Cross- validaion Recall Score: ", cv_score)

    print("\n=============================> CLASSIFICATION REPORT <===================================")

    ## Classification Report
    clf_rep = classification_report(y_test, test_predictions, output_dict=True)

    print(classification_report(y_test, test_predictions,
                                target_names=confusion_labels))


    if verbose:
        print("\n================================> CONFUSION MATRIX <=====================================")

        ## Confusion Matrix HeatMap
        display(confusion_plot(confusion_matrix(y_test, test_predictions),
                               labels=confusion_labels))
        print("\n=======================================> PLOTS <=========================================")


        ## Variable importance plot
        fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))
        roc_axes = axes[0, 0]
        pr_axes = axes[0, 1]
        importances = None

        if importance_plot:
            if not feature_labels:
                raise RuntimeError("'feature_labels' argument not passed "
                                   "when 'importance_plot' is True")

            try:
                importances = pd.Series(clf.feature_importances_,
                                        index=feature_labels) \
                                .sort_values(ascending=False)
            except AttributeError:
                try:
                    importances = pd.Series(clf.coef_.ravel(),
                                            index=feature_labels) \
                                    .sort_values(ascending=False)
                except AttributeError:
                    pass

            if importances is not None:
                # Modifying grid
                grid_spec = axes[0, 0].get_gridspec()
                for ax in axes[:, 0]:
                    ax.remove()   # remove first column axes
                large_axs = fig.add_subplot(grid_spec[0:, 0])

                # Plot importance curve
                feature_importance_plot(importances=importances.values,
                                        feature_labels=importances.index,
                                        ax=large_axs)
                large_axs.axvline(x=0)

                # Axis for ROC and PR curve
                roc_axes = axes[0, 1]
                pr_axes = axes[1, 1]
            else:
                # remove second row axes
                for ax in axes[1, :]:
                    ax.remove()
        else:
            # remove second row axes
            for ax in axes[1, :]:
                ax.remove()


        ## ROC and Precision-Recall curves
        clf_name = clf.__class__.__name__
        roc_plot(y_test, y_probs, clf_name, ax=roc_axes)
        precision_recall_plot(y_test, y_probs, clf_name, ax=pr_axes)

        fig.subplots_adjust(wspace=5)
        fig.tight_layout()
        display(fig)

    ## Dump to report_dict
    dump = dict(clf=clf, train_acc=train_acc, train_time=train_time,
                train_predictions=train_predictions, test_acc=test_acc,
                test_time=test_time, test_predictions=test_predictions,
                test_probs=y_probs, report=clf_rep, roc_auc=roc_auc,
                model_memory=model_mem)

    return clf, dump
```

```python
def compare_models(y_test=None, clf_reports=[], labels=[]):

    ## Classifier Labels
    default_names = [rep['clf'].__class__.__name__ for rep in clf_reports]
    clf_names =  labels if len(labels) == len(clf_reports) else default_names


    ## Compare Table
    table = dict()
    index = ['Train Accuracy', 'Test Accuracy', 'Overfitting', 'ROC Area',
             'Precision', 'Recall', 'F1-score', 'Support']
    for i in range(len(clf_reports)):
        train_acc = round(clf_reports[i]['train_acc'], 3)
        test_acc = round(clf_reports[i]['test_acc'], 3)
        clf_probs = clf_reports[i]['test_probs']
        roc_auc = clf_reports[i]['roc_auc']

        # Get metrics of True Positive class from sklearn classification_report
        true_positive_metrics = list(clf_reports[i]['report']["1"].values())

        table[clf_names[i]] = [train_acc, test_acc,
                               abs(test_acc - train_acc)>0.04, roc_auc] + true_positive_metrics

    table = pd.DataFrame(data=table, index=index)


    ## Compare Plots
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

    # ROC and Precision-Recall
    for i in range(len(clf_reports)):
        clf_probs = clf_reports[i]['test_probs']
        roc_plot(y_test, clf_probs, label=clf_names[i],
                 compare=True, ax=axes[0])
        precision_recall_plot(y_test, clf_probs, label=clf_names[i],
                              compare=True, ax=axes[1])
    # Plot No-Info classifier
    axes[0].plot([0,1], [0,1], linestyle='--', color='green')

    fig.tight_layout()
    plt.close()

    return table.T, fig
```

#  **Revenue Retention Analysis**
Maintaining and expanding your existing customer base is essential for long-term business growth.

We therefore defined a additional evaluation metrics called `Percentage Monthly Revenue Retained` which is defined as,
  
\begin{align}
\text {Revenue Retained (Monthly)} \%=\frac{\sum_{i=1}^{n_{\text {test}}} y_{\text {test}}^{(i)} \times y_{\text {pred}}^{(i)} \times \text {monthlycharges}^{(i)}}{\sum_{i=1}^{n_{\text {test}}} y_{\text {test}}^{(i)} \times \text {monthlycharges}(i)} \times 100
\end{align}

The primary goal of these concepts is to determine which customers have the highest monthly payments each month and whether the models can identify these customers.

`Percentage Total Revenue Retained`,

\begin{align}
\text {Revenue Retained (Total) } \%=\frac{\sum_{i=1}^{n_{\text {test}}} y_{\text {test}}^{(i)} \times y_{\text {pred}}^{(i)} \times \text {totalcharges}^{(i)}}{\sum_{i=1}^{n_{\text {test}}} y_{\text {test}}^{(i)} \times \text {totalcharges}^{(i)}} \times 100
\end{align}  
  
These are revenue-weighted recall scores, similar to Recall in the business world. They represent the revenue that a model retains (or saves) as a result of accurate churn predictions, also known as True Positives.

Similarly, `Revenue Retained per true churn customer (Monthly) ` is defined as,

\begin{align}
\text {Revenue Retained per churn customer (Monthly) }=\frac{\sum_{i=1}^{n_{\text {test}}} y_{\text {test}}^{(i)} \times y_{\text {pred}}^{(i)} \times \text {monthlycharges}^{(i)}}{\sum_{i=1}^{n_{\text {test}}} y_{\text {test}}^{(i)} \times y_{\text {pred}}^{(i)}} \times 100
\end{align}  
  
This metric assesses a classifier's ability to identify Most Valuable Customers (MVC) or high-spending customers. A high value indicates that a classifier identified MVCs better than a lower-valued classifier.  
  
**Note:** The abbreviations of the above formulas are denoted as following:  
> RRM ( % ) => Revenue Retained (Monthly( % )

> RRT ( % )=> Revenue Retained (Total)( % )

> RRM/customer => Revenue Retained (Monthly) per true churn customer

```python
df_charges = list()
new_cols = ['RRM/(customer)', 'RRM(%)', 'RRT(%)']

for rep in report_list:
    y_pred = rep['test_predictions']
    true_positives = (y_test * y_pred).reshape(y_test.shape[0], 1)
    tp_revenue = (x_test_charges * true_positives).sum(axis=0)
    tp_revenue = np.insert(tp_revenue, 0, tp_revenue[0] / true_positives.sum())
    df_charges.append(tp_revenue)

revenue_saved = pd.DataFrame(df_charges, index=clf_labels,
                             columns=new_cols)

compare_table_rev = pd.concat([compare_table, revenue_saved], axis=1)

## True Positive Revenue/ Total Churn Revenue
total_churn_revenue = (x_test_charges * y_test.reshape(y_test.shape[0], 1)).sum(axis=0)
temp_cols = (compare_table_rev.iloc[:, -2:] / total_churn_revenue) * 100
compare_table_rev.iloc[:, -2:] = temp_cols
```
